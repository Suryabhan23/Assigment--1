**HomeSpace** is a furniture company that sells its products across various cities in India. You are provided with the sales details of HomeSpace in the the link below:


url: "https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/HomeSpace%20-%20Data.csv"

**The manager has asked you to provide some insights in the sales of last 3 months. For this please complete the following tasks.**

***Task 1: Import the pandas library.***

```python
import pandas as pd
```

***Task 2: Create a DataFrame named 'hssales' from the given URL and then print the first 5 rows using head() function.***

```python
hssales = pd.read_csv('https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/HomeSpace%20-%20Data.csv')
print(hssales.head())
```

***Task 3: Add a column named "Sales" and calculate the sales by multiplying unit price with units sold.***

```python
hssales['Sales']= hssales['UnitSold'] * hssales['UnitPrice']
hssales.head()
```

***Task 4: Use the "Groupby" function to calculate the total sales for each month and print the resulting data.***

```python
MontlySales= hssales.groupby("Month")['Sales'].sum().reset_index()
MontlySales
```

***Task 5: Use the "Groupby" function to calculate the minimum sales in each city, and print the resulting data.***

```python
CityMinsales= hssales.groupby("City")['Sales'].min()
CityMinsales
```

***Task 6: Use "Groupby" function to determine the maximum sales achieved by each salesperson over the three-month period, and print the resulting data.***

```python
MaxSales= hssales.groupby("Salesperson")['Sales'].max().reset_index()
MaxSales
```

***Task 7: Use the groupby function to calculate the total units sold for each product and print the resulting data.***

```python
Unit_Sold= hssales.groupby('Product')['UnitSold'].sum().reset_index()
Unit_Sold
```

***Task 8: Use the groupby function to calculate the number of unique products sold in each city and print the resulting data.***

```python
Unique_Product_Sold=hssales.groupby('City')['Product'].nunique().reset_index()
Unique_Product_Sold
```

***Task 9: Use the "Groupby" function to calculate the monthly sales made by each salesperson, and print the resulting sales data for each month and salesperson.***

```python
MontlySales= hssales.groupby(["Month","Salesperson"])["Sales"].sum()
MontlySales
```

***Task 10: Use the "Groupby" function to calculate the total units sold for each product in every city, and print the resulting data.***

```python
Product_sold_city= hssales.groupby(["City","Product"])["UnitSold"].sum().reset_index()
Product_sold_city
```

***Task 11: Use the groupby function to calculate the number of unique cities where each salesperson has made sales, and print the resulting data.***

```python
UniqueCity= hssales.groupby("Salesperson")["City"].nunique().reset_index()
UniqueCity
```
