```python
import pandas as pd
import matplotlib.pyplot as plt
Product_Complain = pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/ComplaintsTable.csv")
Product_Complain
Product_Complain.head()
```

```python
 total_complaints= Product_Complain.groupby("ProductCategory")["ComplaintCount"].sum()
 total_complaints
```

```python
sorted_complaints= total_complaints.sort_values(ascending=False)
sorted_complaints
```

```python
 cumulative_sum = sorted_complaints.cumsum()
 cumulative_sum
```

```python
cumulative_percentage=cumulative_sum/sorted_complaints.sum()*100
cumulative_percentage
```

```python
fig,ax1 = plt.subplots()
sorted_complaints.plot(kind='bar',color='C0',ax=ax1)
ax1.set_ylabel('ComplaintCount')

ax2= ax1.twinx()
cumulative_percentage.plot(kind='line',color='C1',ax=ax2,marker='D')
ax2.set_ylabel('Cumulative Percentage %')
```

```python
Top4 = Product_Complain.query('ProductCategory in ["Kitchenware","Bedding","Jewelry","Books"]')
Top4
```

```python
sorted_reasons1 = Top4.groupby("ComplaintReason")["ComplaintCount"].sum().sort_values(ascending=False)
sorted_reasons1
```

```python
cumulative_reason_percentage=round(sorted_reasons1.cumsum()/sorted_reasons1.sum()*100,2)
cumulative_reason_percentage
```

```python
fig, ax1= plt.subplots()
sorted_reasons1.plot(kind='bar',color='C0',ax=ax1)
ax1.set_ylabel('Returns')
ax2=ax1.twinx()
cumulative_reason_percentage.plot(kind='line',color='C1',ax=ax2,marker='D')
ax2.set_ylabel('Cumulative Percentage %')
plt.title('pareto anlysis of the reason associated with return')
```
